```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bokeh.plotting as bp
from bokeh.layouts import gridplot
%matplotlib inline
```

#### Gráfico 1

```python
fig = plt.figure(figsize=(6,4))
plt.title("Chuva acumulada por período no mês de Maio", size=14)
x = range(3)
y = [28.7,1.6,49.8]
plt.axes().spines['right'].set_visible(False)
plt.axes().spines['top'].set_visible(False)
#plt.grid()
plt.grid(axis='y', alpha=0.5, linestyle='--')
plt.xlabel("Período do dia", size=12)
plt.ylabel("Chuva em  mm³", size=12)
plt.xticks(x, (["Manhã", "Tarde", "Noite"]))
plt.bar(x, y, width=0.4, align='center', data=None)
plt.show()
fig.savefig('grafico1.png')
plt.close(fig) #para cerrar la figura generada
```

#### Gráfico 2

```python
fig = plt.figure(figsize=(6,4))
plt.title("Os times que mais venceram o Campeonato Brasileiro da séria A", size=14)
x = range(5)
y = [9,8,7,6,5]
plt.axes().spines['right'].set_visible(False)
plt.axes().spines['top'].set_visible(False)
colors = ['red', 'tan', 'brown',  'pink', 'green']
#plt.grid()
plt.xticks(x, (["Palmeiras", "Santos", "Corinthians", "São Paulo", "Flamengo"]))
plt.xlabel("Times", size=12)
plt.grid(axis='y', alpha=0.5, linestyle='--')
plt.ylabel("Quantidade de vitórias", size=12)
plt.bar(x, y, width=0.6, align='center', data=None,color=colors)
plt.show()
fig.savefig('grafico2.png')
plt.close(fig) 
```

#### Gráfico 3

```python
fig = plt.figure(figsize=(6,4))
plt.title("Temperaturas máximas da última semana em Belém", size=14)
x = range(7)
y = [25.6,27.3,32.0,34.8,30.7,26.5,31.6]
plt.axes().spines['right'].set_visible(False)
plt.axes().spines['top'].set_visible(False)
colors = ['red', 'tan', 'brown','black', 'pink', 'green', 'blue']
plt.grid(axis='y', alpha=0.5, linestyle='--')
plt.xticks(x, (["Domingo", "Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira", "Sábado"]), rotation=15)
plt.xlabel("Días da semana", size=12)
plt.ylabel("Temperatura em °C", size=12)
plt.bar(x, y, width=0.6, align='center', data=None, color=colors)
plt.show()
fig.savefig('grafico3.png')
plt.close(fig) 
```
