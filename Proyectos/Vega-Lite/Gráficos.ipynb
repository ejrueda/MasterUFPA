```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bokeh.plotting as bp
from bokeh.layouts import gridplot
%matplotlib inline
```

#### Gr√°fico 1

```python
fig = plt.figure(figsize=(6,4))
plt.title("Chuva acumulada por per√≠odo no m√™s de Maio", size=14)
x = range(3)
y = [28.7,1.6,49.8]
plt.axes().spines['right'].set_visible(False)
plt.axes().spines['top'].set_visible(False)
#plt.grid()
plt.grid(axis='y', alpha=0.5, linestyle='--')
plt.xlabel("Per√≠odo do dia", size=12)
plt.ylabel("Chuva em  mm¬≥", size=12)
plt.xticks(x, (["Manh√£", "Tarde", "Noite"]))
plt.bar(x, y, width=0.4, align='center', data=None, color="grey")
plt.show()
fig.savefig('grafico1.png')
plt.close(fig) #para cerrar la figura generada
```
