```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig2 = plt.figure(figsize=(3,2))
ax2 = fig2.add_subplot()
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
x = np.arange(-8,8,0.1)
y = 1 / (1 + np.exp(-x))
plt.plot(x, np.ones(len(x))*0.5, linestyle="--", color="grey")
plt.plot(x, y, color="black");
plt.savefig("./sigmoid_function.svg")
```

```python
fig2 = plt.figure(figsize=(3,2))
ax2 = fig2.add_subplot()
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
x = np.arange(-8,8,0.1)
y = np.tanh(x)
plt.plot(x, np.zeros(len(x)), linestyle="--", color="grey")
plt.plot(x, y, color="black");
plt.savefig("./tanh_function.svg")
```
