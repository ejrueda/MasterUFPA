```python
c_power = ["power", "energy", "electric", "electrician", "electricity", "powerlines", "electric", "radiation",
          "reactor", "nuclear"]
c_shake = ["shake", "earthquake", "quake", "tremble", "rubble", "earthquakes", "earthqwakes", "earthqwake", 
           "vibrate", "wobble"]
c_hospital = ["hospital", "hospitals", "die", "dead", "death", "health", "ambulance"]
c_disaster = ["disaster", "disasters"]
c_sewer = ["sewer", "sewage"]
c_evacuation = ["evacuation", "evacuate", "rescue"]
c_build = ["build", "structutral", "bridge", "destroy", "repair"]
others = ["emergency", "dangerous", "problem", "problems", "cry", "apocalypse", "sad", "victims",
          "help", "tragedy", "roads", "scar", "hazard", "damage", "crew", "shudder", "incident", "risk"]
```

#### Para gerar o novo dataset para o grfico por bairro

```python
import ast
data_cluster = pd.DataFrame(columns=["time", "location", "account", "tipo", "word"])
for idx, val in data_pro.iterrows(): #para recorrer os bairros
    for w in ast.literal_eval(val.message): #para trocar o string por uma lista
        if w in c_power:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"power",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in c_shake:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"shake",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in c_hospital:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"hospital",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in c_disaster:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"disaster",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in c_sewer:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"sewer",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in c_evacuation:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"rescue",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in c_build:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"build",w]],
                                                           columns=["time","location","account","tipo","word"]))
        if w in others:
            data_cluster = data_cluster.append(pd.DataFrame([[idx,val.location,val.account,"others",w]],
                                                           columns=["time","location","account","tipo","word"]))
```

```python
data_cluster = data_cluster.set_index("time") #set index
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook, reset_output, curdoc
from bokeh.embed import file_html
from bokeh.models import LabelSet, Label, ColumnDataSource, FactorRange, HoverTool
from bokeh.models.annotations import Title
from bokeh.transform import factor_cmap
import bokeh.plotting as bp
from bokeh.layouts import row, column,layout, gridplot
from bokeh.models.widgets import Slider, TextInput, RangeSlider, Select
from bokeh.transform import cumsum, jitter
from bokeh.palettes import Category10, Category20

#os conjuntos de dados so carrregados
data_pro = pd.read_csv("./grafico2/data/data_geral_normalizada.csv", index_col=0, parse_dates=True,
                       infer_datetime_format=True)
data_cluster = pd.read_csv("./grafico2/data/data_words.csv", index_col=0, infer_datetime_format=True,
                           parse_dates=True)
#bairros
bairros = ["Palace Hills", "Northwest", "Old Town", "Safe Town", "Southwest", "Downtown",
         "Wilson Forest", "Scenic Vista", "Broadview", "Chapparal", "Terrapin Springs",
         "Pepper Mill", "Cheddarford", "Easton", "Weston", "Southton", "Oak Willow",
         "East Parton", "West Parton"]
bairro_init = "Palace Hills"
#grfico, palavras chaves nos bairros
p1 = figure(plot_width=1150, plot_height=500, y_range=bairros, x_axis_type='datetime',
           title="palavras chave por bairro no tempo",
          tools="hover,pan,wheel_zoom,box_zoom,reset,save",
          background_fill_color="#F7F7F7")
source1 = ColumnDataSource(data_pro)
p1.circle(x='time', y=jitter('location', width=0.7, range=p1.y_range),  source=source1, alpha=0.3)
p1.hover.tooltips = [("palavras", "@message")]
p1.x_range.range_padding = 0.01
p1.ygrid.grid_line_color = None

#grfico dois, cluster das palavras chave
data_aux = data_cluster.loc[data_cluster.location==bairro_init]
source2 = ColumnDataSource(data_aux)
pvras = ["power","shake","hospital","disaster","sewer","rescue","build","others"]
p2 = figure(plot_width=1200, plot_height=400, y_range=pvras, x_axis_type='datetime',
           title="Palavras chave por bairro",
          tools="hover,pan,wheel_zoom,box_zoom,reset,save",
          background_fill_color="#F7F7F7", x_range=p1.x_range)
p2.circle(x='time', y=jitter('tipo', width=0.6, range=p2.y_range),  source=source2, alpha=0.5,)
p2.hover.tooltips = [("datetime", "@time{%F %T}"),
                    ("tipo", "@tipo"),
                    ("palavra", "@word")]
p2.hover.formatters = {'time': 'datetime'}
p2.x_range.range_padding = .01
p2.ygrid.grid_line_color = None
p2.xaxis.axis_label = "dias do terremoto"
p2.yaxis.axis_label = "palavras chave no terremoto"

#Barras de interao
select_vec = Select(title="Bairro:", value=bairro_init, options=bairros)

inputs = column([select_vec], width=150)
row_1 = row([inputs, p1])
row_2 = row([p2])

curdoc().add_root(row(gridplot([[row_1],[row_2]]), width=400))
```
