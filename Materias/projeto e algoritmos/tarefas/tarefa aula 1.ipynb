## Questaõ 2:
- Implemente em uma linguagem de programação a sua escolha o algoritmo abaixo que verifica se o valor de entrada é, ou não, um número primo. Existe pior e melhor caso?

```python
import numpy as np
def primo(n):
    """
    Dize se o número é primo ou não
    """
    j = 2
    while(j < n and np.mod(n,j) != 0):
        j += 1
    if (j==n):
        print("O número é Primo")
    else:
        print("O número não é Primo")
```

```python
primo(10007)
```

### Melhor e pior caso:
- o melhor caso e quando o primo é o número 2, aí ele fica com ordem: $$\Omega(1)$$ <br>
- e o pior caso é quando o número é primo e é maior, nesse caso o ordem fica assim: $$O(n)$$

## Questão 3:
- Implemente em uma linguagem de programação a sua escolha o algoritmo abaixo que lista os números primos menores ou igual ao valor de entrada. Existe pior e melhor caso?

```python
def lista_primo(n):
    """
    lista os primos menores que "n"
    """
    A = []
    for i in range(2,n+1):
        j = 2
        while(j < i and np.mod(i,j) != 0):
            j += 1
        if (j==i):
            A.append(i)
    return A
```

```python
n = 60
print("n = ",n," | números primos menores que ",n,":")
print(lista_primo(n))
```

### Melhor e pior caso:
- o melhor caso e quando o n seja o número 2, aí ele fica com ordem:
$$T(n)=C_{5}+C_{6}\cdot2+C_{7}+C_{8}+C_{10}+C_{11}$$
$$\Omega(2)$$
<br>
- e o pior caso é quando o número é primo e é maior, nesse caso o ordem fica assim: 
$$T(n)=C_{5}+C_{6}\cdot(n)\cdot[C_{7}\cdot(n-1)+C_{8}\cdot(n-1)+C_{9}\cdot(n-2)+C_{10}\cdot(n-1)+C_{11}\cdot x]\,onde\,x\,é\,menor\,que\,n$$
$$O(n^{2})$$

## Questão 4 
- Cada um dos algoritmos abaixo recebe um inteiro positivo e devolve outro inteiro positivo. Os dois algoritmos devolvem o mesmo número se receberem o mesmo valor de entrada n? Qual dos dois algoritmos é mais eficiente?

```python
def soma_quadrados_A(n):
    x = 0
    for j in range(1,n+1):
        x = x + (j*j)
    return x
```

```python
def soma_quadrados_B(n):
    x = n*(n+1)*(2*n+1)
    x = x/6
    return x
```

- Os dois algoritmos devolvem o mesmo número, o qual é a sumatoria dos primeiros n quadrados.

```python
n = 10
print("n: ", n)
print("soma quadrados A: ",soma_quadrados_A(n))
print("soma_quadrados_B: ",soma_quadrados_B(n))
```

- O algoritmo mais eficiente é o algoritmo B, como a imagem de abaixo mostra:

- Compare assintoticamente o tempo de execução (apresente dados, tabelas e gráficos) dos algoritmos de ordenação:
    - Inserção;
    - BubbleSort; e
    - Seleção.

```python
def insertionSort(arr): 
    for i in range(1, len(arr)): 
        key = arr[i] 
        j = i-1
        while j >=0 and key < arr[j]: 
                arr[j+1] = arr[j] 
                j -= 1
        arr[j+1] = key 
    #return arr
```

```python
insertionSort(np.random.randint(1,100,size=(10)))
```

```python
def bubbleSort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]         
    #return arr
```

```python
bubbleSort(np.random.randint(1,100,size=(10)))
```

```python
def selectionSort(A):
    for i in range(len(A)): 
        min_idx = i 
        for j in range(i+1, len(A)): 
            if A[min_idx] > A[j]: 
                min_idx = j       
        A[i], A[min_idx] = A[min_idx], A[i]
    #return A
```

```python
selectionSort(np.random.randint(1,100,size=(10)))
```

```python
t_is = []
t_bs = []
t_ss = []
for i in range(1,1000):
    x = np.random.randint(low=0, high=100,size=i*5)
    t_i = time()
    insertionSort(x)
    t_f = time()
    t_is.append(t_f-t_i)
    ###############
    t_i = time()
    bubbleSort(x)
    t_f = time()
    t_bs.append(t_f-t_i)
    ###############
    t_i = time()
    selectionSort(x)
    t_f = time()
    t_ss.append(t_f-t_i)
    print("[",i,"]/[",1000,"]")
```
