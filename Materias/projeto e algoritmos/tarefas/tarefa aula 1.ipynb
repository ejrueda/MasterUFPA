## Questão 1:
- Qual é o menor valor de entrada n (considere n > 0) tal que um algoritmo cujo tempo de execução é 10n² é mais rápido que um algoritmo cujo tempo de execução é 2^n ? Qual desses algoritmos você considera mais eficiente?

```python
x1 = [10*(i**2) for i in range(1,20)]
x2 = [2**i for i in range(1,15)]
```

```python
plt.title("tempo de execução (n)", fontsize=15)
plt.ylabel("tempo de execução" , fontsize=13)
plt.xlabel("n", fontsize=13)
plt.plot(range(len(x1)), x1, label="10n²")
plt.plot(range(len(x2)), x2, label="2^n", color="red")
#plt.yscale("log")
plt.xticks(range(len(x1)),range(1,len(x1)+1))
plt.grid()
plt.legend();
```

```python
print("O menor valor para que 10n² seja melhor que o 2^n é n=10")
print("---------------------------------------")
print("T(n)=10n² = ", x1[9])
print("T(n)=2^n = ", x2[9])
print("---------------------------------------")
print("O algoritmo mais eficente assintoticamente é 10n²")
```

## Questaõ 2:
- Implemente em uma linguagem de programação a sua escolha o algoritmo abaixo que verifica se o valor de entrada é, ou não, um número primo. Existe pior e melhor caso?

```python
import numpy as np
def primo(n):
    """
    Dize se o número é primo ou não
    """
    j = 2
    while(j < n and np.mod(n,j) != 0):
        j += 1
    if (j==n):
        print("O número é Primo")
    else:
        print("O número não é Primo")
```

```python
primo(10007)
```

### Melhor e pior caso:
- o melhor caso e quando o primo é o número 2, aí ele fica com ordem: $$\Omega(1)$$ <br>
- e o pior caso é quando o número é primo e é maior, nesse caso o ordem fica assim: $$\theta(n)$$
