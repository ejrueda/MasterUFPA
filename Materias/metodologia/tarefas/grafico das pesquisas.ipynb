```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
year = [2014,2015,2016,2017,2018, 2019] 
q = np.array([173, 158, 171, 156, 146, 159])
plt.figure(figsize=(10,5))
plt.title("Estudos por ano", fontsize=16)
plt.xlabel("Ano", fontsize=14)
plt.ylabel("Quantidade", fontsize=14)
plt.grid(axis="y")
plt.plot(year, q)
plt.scatter(year, q);
```
